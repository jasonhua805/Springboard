# Data Cleaning

First, let's translate the column names into English.

```python
new_columns = ["Date","ID","Employee","Country","Sex","Age","Account Opened","Opened In Last Six Months","Months","Primary Customer","Last Date As Primary Customer","Customer Type","Customer Relation",
                  "Spanish Resident","Foreigner","Employee Spouse","Channel","Deceased","Address Type","Province Code","Province Name","Activity Index","Gross Household Income","Segmentation","Savings Account","Guarantees",
                  "Current Account", "Derivatives Account", "Payroll Account", "Junior Account", "More Particular Account","Particular Account","Particular Plus Account", "Short Term Deposits","Medium Term Deposits",
                  "Long Term Deposits","eAccount","Funds","Mortgage","Pensions","Loans","Taxes","Credit Card","Securities","Home Account","Payroll","Nominal Pensions","Direct Debit"]
mapping = dict(zip(list(df.columns),new_columns))
df.rename(columns = mapping, inplace=True)
```

This dataset is absolutely enormous. Let's pull out a sample of it to perform our EDA on to reduce processing demand.

```python
ids = pd.Series(df["ID"].unique())
sample_ids = ids.sample(n=10000)
dfs = df[df.ID.isin(sample_ids)]
```

```python
dfs.info()
```

A lot of customers who have missing data in some of the identifying information (e.g. Sex, Employee, Country) seem to be missing data for all of the fields. Since this dataset is already so large - it should be safe to drop these customers.

```python
dfs = dfs[dfs['Account Opened'].notna()]
dfs.info()
```

```python
dfs.info()
```

Address Type has only one value and it is a meaningless column. We also don't need a Province Code column if we're already going to have Province Name. Let's get rid of them now.

```python
dfs = dfs.drop(columns=['Address Type','Province Code'])
```

```python
# Most customers are classified as a "Primary Customer" so let's fill that in for the Customer Type column.
# For customer relation, there is a roughly equal distribution of both "inactive" and "active" customers. Let's forward fill in the missing values.
dfs.loc[dfs['Customer Type'].isna(),'Customer Type'] = 1
dfs['Customer Relation'].fillna(method='ffill',inplace=True)
```

```python
dfs['Channel'].value_counts()
```

Because there is no reliable way to fill in missing values for "Channel" we'll make an "Unknown" value to fill.

```python
dfs['Channel'].fillna(value='Unknown',inplace=True)
```
